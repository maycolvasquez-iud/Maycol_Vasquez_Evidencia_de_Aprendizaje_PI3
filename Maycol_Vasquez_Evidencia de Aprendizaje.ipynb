## **Importación de librerías y carga de datos**

```python
# Se importan librerias necesarias
import pandas as pd
import numpy as np
from ydata_profiling import ProfileReport
from google.colab import files
```

```python
# Se carga el dataset
ruta_archivo = '/content/drive/MyDrive/IUDIGITAL/Ingeniería de software y datos/Semestre 2/Bloque 2/Proyecto Integrado III Analitica de datos/Actividad/Telco-Customer-Churn.csv'
df = pd.read_csv(ruta_archivo)
```

```python
# Se muestran las primeras filas para verificar que cargó correctamente
df.head()
```

## **Limpieza rápida**

```python
# Se corregi el problema detectado en la exploración:
# TotalCharges tiene espacios en blanco " " en lugar de nulos reales.
# Lo forzamos a numérico; los espacios se convertirán en NaN.
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Verificamos los tipos de datos para confirmar que TotalCharges ahora es float64
df.dtypes
```

## **Generación y visualización del reporte**

```python
# See gnera el reporte de Profiling
profile = ProfileReport(df, title="Reporte de EDA - Telco Customer Churn", explorative=True)

# Se muestra el reporte de forma interactiva dentro de Colab
profile.to_notebook_iframe()
```

## 5. Distribuciones

El reporte de profiling revela comportamientos asimétricos y desbalances cruciales:

- **Distribución del Target (Churn):**

- **Distribución de tenure:**

## 6. Correlaciones

Validando la matriz de correlación se tiene:

- **Correlación positiva fuerte:**  
  Existe una correlación casi perfecta entre **tenure** y **TotalCharges** (a mayor tiempo, mayor cobro acumulado), y entre **MonthlyCharges** y **TotalCharges**.

- **Correlación con el Churn:**
  - **Negativa con tenure:**  
    Los clientes con menor antigüedad (especialmente los de 1 a 6 meses) tienen una probabilidad muchísimo más alta de hacer Churn.  

  - **Positiva con Contract:**  
    El contrato de mes a mes ("Month-to-Month") está altamente correlacionado con el abandono.  
    Los contratos de 1 o 2 años casi anulan la probabilidad de Churn.  

  - **Positiva con InternetService:**  
    Los clientes con servicio de "Fiber optic" (fibra óptica) presentan tasas de abandono significativamente superiores a los de DSL, probablemente vinculado a costos más altos o fallas en el servicio.
